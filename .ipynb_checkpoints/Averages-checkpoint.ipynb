```python
import numpy as np
import pandas as pd
import scipy.stats as s
import seaborn as sns
```

```python
colors=['#7D110C', '#cdc4b6']
```

```python
data = np.random.normal(100, 20, 47)
data = np.append(data, [500, 502, 498])
```

```python
arithmetic_mean = np.mean(data)
harmonic_mean = s.hmean(data)
print(f'Arithemtic Mean = {arithmetic_mean:.2f}')
print(f'Harmonic Mean - {harmonic_mean:.2f}')
```

```python
sns.set_palette(colors)
chart = sns.histplot(data, kde=False)
```

```python
observed_std = np.std(data)
```

```python
x = np.linspace(arithmetic_mean - 3 * observed_std, arithmetic_mean + 3 * observed_std)
arithmetic_y = s.norm.pdf(x, loc=arithmetic_mean, scale=observed_std)
harmonic_y = s.norm.pdf(x, loc=harmonic_mean, scale=observed_std)

df = pd.DataFrame({'x':x, 'y':arithmetic_y, 'Type of Mean':'Arithmetic'})
df = df.append(pd.DataFrame({'x':x, 'y':harmonic_y, 'Type of Mean':'Harmonic'}))
```

```python
# sns.set_palette(colors)
chart = sns.lineplot(data=df, x=df['x'], y=df['y'], hue=df['Type of Mean'])
chart.axvline(arithmetic_mean, color=colors[0])
chart.axvline(harmonic_mean, color=colors[1])
```
